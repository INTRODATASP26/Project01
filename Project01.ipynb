## Load Data

Load your data into `Python` here.

```python
import json
import os

data_folder = 'data/Spotify Extended Streaming History'
all_data = []

for filename in os.listdir(data_folder):
    if filename.endswith('.json'):
        filepath = os.path.join(data_folder, filename)
        with open(filepath, 'r', encoding='utf-8') as file:
            data = json.load(file)
            all_data.extend(data)  # Combine all JSON arrays into one list
            print(f"Loaded {filename}: {len(data)} records")

print(f"\nTotal records: {len(all_data)}")

# Convert to DataFrame for easier analysis
streaming_df = pd.DataFrame(all_data)
print(f"\nDataFrame shape: {streaming_df.shape}")
print(f"\nColumns: {streaming_df.columns.tolist()}")
print(f"\nFirst row:\n{streaming_df.iloc[0]}")
```

## Explore Data

Use the libraries and commands we've seen, as well as the `DataFrame` object to navigate your data.

```python
print(f"Total records: {len(all_data)}")

# Create df with time, track name, artist name, album name, and country I listened to music in
streaming_df = pd.DataFrame(all_data)[['ts', 'platform', 'ms_played', 'conn_country', 'ip_addr', 'master_metadata_track_name', 'master_metadata_album_artist_name', 'master_metadata_album_album_name', 'spotify_track_uri', 'episode_name', 'episode_show_name', 'spotify_episode_uri', 'audiobook_title', 'audiobook_uri', 'audiobook_chapter_uri', 'audiobook_chapter_title', 'reason_start', 'reason_end', 'shuffle', 'skipped', 'offline', 'offline_timestamp', 'incognito_mode']]

print(f"DataFrame shape: {streaming_df.shape}")

# Display the full DataFrame as a spreadsheet table
streaming_df
```

## Show Data

Show some interesting findings.

```python
#ps: I did most of the explorations by myseld but I asked Claude for help when I got stuck :)

#How much I spent listening to music each day per year? 

# Convert ts to datetime
streaming_df['ts'] = pd.to_datetime(streaming_df['ts'])

# Extract year
streaming_df['year'] = streaming_df['ts'].dt.year

# Group by year and sum milliseconds played
yearly_ms = streaming_df.groupby('year')['ms_played'].sum()

# Convert milliseconds to hours
yearly_hours = yearly_ms / (1000 * 60 * 60)

# Calculate daily average (hours per day in that year)
daily_avg = yearly_hours / 365

# Create a summary DataFrame
summary = pd.DataFrame({
    'Year': yearly_hours.index,
    'Total Hours': yearly_hours.values.round(2),
    'Daily Average (hours)': daily_avg.values.round(2),
    'Daily Average (minutes)': (daily_avg * 60).values.round(0)
})

print("=" * 60)
print("ANNUAL AND DAILY LISTENING STATISTICS")
print("=" * 60)
summary

# Create line graph for daily average per year
plt.figure(figsize=(10, 6))
plt.plot(summary['Year'], summary['Daily Average (hours)'], marker='o', linewidth=2.5, markersize=8, color='#1DB954')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Daily Average (hours)', fontsize=12, fontweight='bold')
plt.title('Average Daily Listening Time Per Year', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(summary['Year'])

# Add data labels on each point
for i, row in summary.iterrows():
    plt.text(row['Year'], row['Daily Average (hours)'] + 0.1, f"{row['Daily Average (hours)']:.2f}h", 
             ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Extract year from ts column
streaming_df['ts'] = pd.to_datetime(streaming_df['ts'])
streaming_df['year'] = streaming_df['ts'].dt.year

# Top 3 artists per year
print("=" * 80)
print("TOP 3 ARTISTS PER YEAR")
print("=" * 80)

top_artists = streaming_df.groupby(['year', 'master_metadata_album_artist_name'])['ms_played'].sum().reset_index()
top_artists.columns = ['Year', 'Artist', 'Total MS']
top_artists['Hours'] = (top_artists['Total MS'] / (1000 * 60 * 60)).round(2)
top_artists = top_artists.sort_values(['Year', 'Total MS'], ascending=[True, False])

# Display each year separately
for year in sorted(top_artists['Year'].unique()):
    year_data = top_artists[top_artists['Year'] == year].head(3).copy()
    year_data['Rank'] = range(1, len(year_data) + 1)
    year_data = year_data[['Rank', 'Artist', 'Hours']]
    print(f"\n{year}")
    print(year_data.to_string(index=False))
```

```python
# Top 3 songs per year

print("\n" + "=" * 80)
print("TOP 3 SONGS PER YEAR")
print("=" * 80)

top_songs = streaming_df.groupby(['year', 'master_metadata_track_name', 'master_metadata_album_artist_name'])['ms_played'].sum().reset_index()
top_songs.columns = ['Year', 'Song', 'Artist', 'Total MS']
top_songs['Hours'] = (top_songs['Total MS'] / (1000 * 60 * 60)).round(2)
top_songs = top_songs.sort_values(['Year', 'Total MS'], ascending=[True, False])

# Display each year separately
for year in sorted(top_songs['Year'].unique()):
    year_data = top_songs[top_songs['Year'] == year].head(3).copy()
    year_data['Rank'] = range(1, len(year_data) + 1)
    year_data = year_data[['Rank', 'Song', 'Artist', 'Hours']]
    print(f"\n{year}")
    print(year_data.to_string(index=False))
```

```python
#How did Stray Kids ruined my life?
#context: I am really obsessed with kpop group Stray Kids. This is an autopsy.

# Filter track data for Stray Kids
stray_kids_data = streaming_df[streaming_df['master_metadata_album_artist_name'] == 'Stray Kids'].copy()

# Group by year and sum milliseconds
yearly_stray_kids = stray_kids_data.groupby('year')['ms_played'].sum().reset_index()
yearly_stray_kids.columns = ['Year', 'Total MS']
yearly_stray_kids['Hours'] = (yearly_stray_kids['Total MS'] / (1000 * 60 * 60)).round(2)
yearly_stray_kids['Streams'] = stray_kids_data.groupby('year').size().values

print("=" * 70)
print("STRAY KIDS DOMINATION CHART")
print("=" * 70)
print(yearly_stray_kids[['Year', 'Streams', 'Hours']].to_string(index=False))

# Create line graph
plt.figure(figsize=(10, 6))
plt.plot(yearly_stray_kids['Year'], yearly_stray_kids['Hours'], marker='o', linewidth=2, markersize=8, color='#FF006E')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Hours Listened', fontsize=12)
plt.title('Stray Kids Dominating My Life: a chart', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(yearly_stray_kids['Year'])
for i, row in yearly_stray_kids.iterrows():
    plt.text(row['Year'], row['Hours'] + 1, f"{row['Hours']}h", ha='center', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# 1. What was the first Stray Kids song I ever listened to?
print("=" * 70)
print("FIRST STRAY KIDS SONG")
print("=" * 70)
first_stray_kids = stray_kids_data.sort_values('ts').iloc[0]
print(f"Song: {first_stray_kids['master_metadata_track_name']}")
print(f"Date: {first_stray_kids['ts']}")

# 2. Most listened Stray Kids song
print("\n" + "=" * 70)
print("MOST LISTENED STRAY KIDS SONG")
print("=" * 70)
most_listened_sk = stray_kids_data.groupby('master_metadata_track_name')['ms_played'].sum().idxmax()
most_listened_sk_hours = stray_kids_data.groupby('master_metadata_track_name')['ms_played'].sum().max() / (1000 * 60 * 60)
print(f"Song: {most_listened_sk}")
print(f"Total Hours: {most_listened_sk_hours:.2f}h")

# 3. Monthly average for Stray Kids in 2025
print("\n" + "=" * 70)
print("STRAY KIDS DOMINATED MY 2025: MONTHLY BREAKDOWN")
print("=" * 70)
stray_kids_2025 = stray_kids_data[stray_kids_data['year'] == 2025].copy()
stray_kids_2025['month'] = pd.to_datetime(stray_kids_2025['ts']).dt.month
monthly_2025 = stray_kids_2025.groupby('month')['ms_played'].sum().reset_index()
monthly_2025.columns = ['Month', 'Total MS']
monthly_2025['Hours'] = (monthly_2025['Total MS'] / (1000 * 60 * 60)).round(2)

# Create a complete list of all 12 months
all_months = pd.DataFrame({'Month': range(1, 13)})
monthly_2025_full = all_months.merge(monthly_2025, on='Month', how='left')
monthly_2025_full['Hours'] = monthly_2025_full['Hours'].fillna(0)
monthly_2025_full['Month Name'] = monthly_2025_full['Month'].apply(lambda x: pd.Timestamp(year=2025, month=x, day=1).strftime('%B'))

for i, row in monthly_2025_full.iterrows():
    print(f"  • {row['Month Name']}: {row['Hours']}h")
    
print(f"\nMonthly Average: {monthly_2025_full['Hours'].mean():.2f}h per month")
```
